```python
#### Fly Head-Direction Ring Attractor Model

### Connectivity Parameters

## Neurons

# EB-PB-Gall (EPG) neurons
N_EPG_L = 9 # number of neurons
N_EPG_R = 9 # number of neurons
N_EPGs = N_EPG_L + N_EPG_R

# PB-EB-NO (PENa) neurons
N_PENa_L = 8 # number of neurons
N_PENa_R = 8 # number of neurons
N_PENs = N_PENa_L + N_PENa_R

# Ring Neurons
N_R = 10 # number of neurons

# Delta Neurons
N_D = 8 # number of neurons

# GLNO neurons
N_GLNO_R = 1 # number of neurons
N_GLNO_L = 1 # number of neurons
N_GLNOs = N_GLNO_R + N_GLNO_L

# Total number of neurons
N = N_EPGs + N_PENs + N_R + N_D + N_GLNOs

## Excitatory Connectivity matrix
C_ACh = np.zeros((N,N))

EPG_scale = 1

# EPG to EPG self excitation (sinusoidal decay)
for i in range(N_EPGs):
    for j in range(N_EPGs):
        if abs(i-j) < EPG_scale:
            C_ACh[i, j] = (1+np.cos(2*np.pi*(i-j)/EPG_scale))/2

# EPG to PENa excitation
for n,i in enumerate(range(N_EPG_L)):
    for m,j in enumerate(range(N_EPGs, N_EPGs+N_PENa_L)):
        if n==m:
            C_ACh[i, j] = 1
for n,i in enumerate(range(N_EPGs-1, N_EPGs-1-N_EPG_R, -1)):
    for m,j in enumerate(range(N_EPGs+N_PENs-1, N_EPGs+N_PENs-1-N_PENa_R, -1)):
        if n==m:
            C_ACh[i, j] = 1

# EPG to Ring excitation
for i in range(N_EPGs):
    for j in range(N_EPGs + N_PENs, N_EPGs + N_PENs + N_R):
        C_ACh[i, j] = 1

# EPG to Delta excitation (sinusoidal)
for i in range(N_EPGs):
    for n,j in enumerate(range(N_EPG_L + N_EPG_R + N_PENa_L + N_PENa_R + N_R, N_EPG_L + N_EPG_R + N_PENa_L + N_PENa_R + N_R + N_D)):
        C_ACh[i, j] = (1 + np.sin(2*np.pi*n/N_D + 2*np.pi*i/N_EPG_L + np.pi))/2


# PENa_L to EPG excitation (left shift)
for n,i in enumerate(range(N_EPGs,N_EPGs+N_PENa_L)):
    for m,j in enumerate(range(N_EPG_L)):
        if m-n == 1:
            C_ACh[i, j] = 2
        elif m-n == 2:
            C_ACh[i, j] = 1

for n,i in enumerate(range(N_EPGs+N_PENa_L-1,N_EPGs-1,-1)):
    for m,j in enumerate(range(N_EPGs-1, N_EPGs-1-N_EPG_R, -1)):
        if m-n == 0:
            C_ACh[i, j] = 2
        elif m-n == 1:
            C_ACh[i, j] = 1

# PENa_R to EPG excitation (right shift)
for n,i in enumerate(range(N_EPGs+N_PENs-1,N_EPGs+N_PENs-1-N_PENa_R, -1)):
    for m,j in enumerate(range(N_EPGs-1, N_EPGs-1-N_EPG_R, -1)):
        if m-n == 1:
            C_ACh[i, j] = 2
        elif m-n == 2:
            C_ACh[i, j] = 1
            
for n,i in enumerate(range(N_EPGs+N_PENa_L,N_EPGs+N_PENa_L+N_PENa_R)):
    for m,j in enumerate(range(N_EPG_L)):
        if m-n == 0:
            C_ACh[i, j] = 2
        elif m-n == 1:
            C_ACh[i, j] = 1

# Atypical PENa to EPG excitation
for i in range(N_EPGs+N_PENa_L-1, N_EPGs+N_PENa_L+1):
    for j in range(N_EPGs):
        if j==N_EPG_L-1 or j==N_EPG_L:
            C_ACh[i, j] = 2
        if j==0 or j==N_EPGs-1:
            C_ACh[i, j] = 3
        if j==1 or j==N_EPGs-2:
            C_ACh[i, j] = 1

# Excitatory Connectivity matrix (NMDA)
C_NMDA = np.zeros((N,N))

# GLNO to PENa excitation
for i in range(N_EPGs+N_PENs+N_R+N_D, N_EPGs+N_PENs+N_R+N_D+N_GLNO_L):
    for j in range(N_EPGs+N_PENa_L, N_EPGs+N_PENa_L+N_PENa_R):
        C_NMDA[i, j] = 1
for i in range(N_EPGs+N_PENs+N_R+N_D+N_GLNO_L, N_EPGs+N_PENs+N_R+N_D+N_GLNO_L+N_GLNO_R):
    for j in range(N_EPGs, N_EPGs+N_PENa_L):
        C_NMDA[i, j] = 1

## Inhibitory Connectivity matrix
C_GABA = np.zeros((N,N))

# Ring to EPG inhibition
for i in range(N_EPGs+N_PENs, N_EPGs+N_PENs+N_R):
    for j in range(N_EPGs):
        C_GABA[i, j] = 1

# Delta to EPG inhibition (sinusoidal)
for i in range(N_EPGs+N_PENs+N_R, N_EPGs+N_PENs+N_R+N_D):
    for n,j in enumerate(range(N_EPGs)):
        C_GABA[i, j] = ((1 + np.sin(2*np.pi*n/N_D + 2*np.pi*i/N_EPG_L))/2)**4

# lateral inhibition between GLNOs
for i in range(N_EPGs+N_PENs+N_R+N_D, N_EPGs+N_PENs+N_R+N_D+N_GLNOs):
    for j in range(N_EPGs+N_PENs+N_R+N_D, N_EPGs+N_PENs+N_R+N_D+N_GLNOs):
        if i!=j:
            C_GABA[i, j] = 1

# plot connectivity matrices
plot_connectivity(C_ACh, C_GABA, C_NMDA)

# Base Scale Factor

G_EPG_EPG = 0.1
G_EPG_PENa = 0.3
G_EPG_R = 0.05
G_EPG_D = 0.05
G_PENa_EPG = 0
G_GLNO_PENa = 0

G_R_EPG = 1.5/N_R
G_D_EPG = 1.5
G_GLNO_GLNO = 0

# Scale connectivity matrices
C_ACh[:N_EPGs, :N_EPGs] *= G_EPG_EPG
C_ACh[:N_EPGs, N_EPGs:N_EPGs+N_PENs] *= G_EPG_PENa
C_ACh[:N_EPGs, N_EPGs+N_PENs:N_EPGs+N_PENs+N_R] *= G_EPG_R
C_ACh[:N_EPGs, N_EPGs+N_PENs+N_R:N_EPGs+N_PENs+N_R+N_D] *= G_EPG_D
C_ACh[N_EPGs:N_EPGs+N_PENs, :N_EPGs] *= G_PENa_EPG

C_GABA[N_EPGs+N_PENs:N_EPGs+N_PENs+N_R, :N_EPGs] *= G_R_EPG
C_GABA[N_EPGs+N_PENs+N_R:N_EPGs+N_PENs+N_R+N_D, :N_EPGs] *= G_D_EPG
C_GABA[N_EPGs+N_PENs+N_R+N_D:, N_EPGs+N_PENs+N_R+N_D:] *= G_GLNO_GLNO

C_NMDA[N_EPGs+N_PENs+N_R+N_D:, N_EPGs:N_EPGs+N_PENs] *= G_GLNO_PENa

# weight noise
W_noise = 0.02 # percentage of weight
C_ACh += W_noise * np.random.randn(N,N) * C_ACh
C_GABA += W_noise * np.random.randn(N,N) * C_GABA

# Transpose
C_ACh = C_ACh.T
C_GABA = C_GABA.T
```

```python
## LIF model parameters
Cm = 0.1 # nF (membrane capacitance)
tau_m = 15 # ms (membrane time constant)
gL = Cm/tau_m # nS (leak conductance)
VL = -70 # mV (leak reversal potential)
Vth = -45 # mV (spike threshold)
Vreset = -75 # mV (refractory reset)
tau_ref = 1 # ms (refractory period)

## Synaptic parameters
# conductances
gGABA = 0.1 # nS (GABA conductance)
gACh = 0.1 # nS (ACh conductance)
gNMDA = 0.1 # nS (NMDA conductance)

# time constants
tau_gaba = 5 # ms (GABA synaptic time constant)
tau_ach = 20 # ms (ACh synaptic time constant)
tau_nmda = 100 # ms (NMDA synaptic time constant)

# reversal potentials
VGABA = -70 # mV (GABA reversal potential)
VACh = 0 # mV (ACh reversal potential)
VNMDA = 0 # mV (NMDA reversal potential)

# activation rates
alpha_nmda = 0.6332 # (NMDA activation rate)

# magnesium concentration
Mg = 1 # mM (magnesium concentration)

# synaptic noise
sigma_s = 0.5 # nA (synaptic noise)

# simulation parameters
dt = 0.1 # ms (simulation time step)
T = 1000 # ms (simulation time)
t = np.arange(0, T, dt) # ms (simulation time vector)

refractory_update = (VGABA-Vreset)/(tau_ref/dt) # (refractory update)

# experimental parameters
Vs = VL * np.ones((N, len(t))) # mV (membrane potential)
Is = np.zeros((N, len(t))) # nA (membrane current)
Ss = np.zeros((N, len(t))) # binary (spike train)
Sgs = np.zeros((N, N, len(t))) # GABA synaptic gating variable
Sas = np.zeros((N, N, len(t))) # ACh synaptic gating variable
Sns = np.zeros((N, N, len(t))) # NMDA synaptic gating variable

# Injection vectors
I_EPGv, I_PEN_Lv, I_PEN_Rv, I_Rv, I_Dv = np.zeros((5, N))
I_EPGv[:N_EPGs] = 1
I_PEN_Lv[N_EPGs:N_EPGs+N_PENa_L] = 1
I_PEN_Rv[N_EPGs+N_PENa_L:N_EPGs+N_PENa_L+N_PENa_R] = 1
I_Rv[N_EPGs+N_PENa_L+N_PENa_R:N_EPGs+N_PENa_L+N_PENa_R+N_R] = 1
I_Dv[N_EPGs+N_PENa_L+N_PENa_R+N_R:] = 1

def lif_update(V,Sg,Sa,Sn,S,I):

    # neuron currents
    I_leak = gL * (VL - V) # nA (leak current)
    I_gaba = gGABA * (VGABA - V) * np.sum(Sg * C_GABA, axis=1) # nA (GABA current)
    I_ACh = gACh * (VACh - V) * np.sum(Sa * C_ACh, axis=1) # nA (ACh current)
    I_nmda = gNMDA * (VNMDA - V)/(1 + (Mg/3.57 * np.exp(-0.062 * V))) * np.sum(Sn * C_NMDA, axis=1) # nA (NMDA current)
    I_net = I_leak + I_gaba + I_ACh + np.random.normal(0, sigma_s, N) # nA (total current)

    # membrane potential update (conditional for refractory period where V<VGABA)
    V[V<VGABA] += refractory_update
    V[V>=VGABA] += ((I*I_EPGv + I_net) / Cm * dt)[V>=VGABA] # mV (membrane potential)
    
    # synaptic gating variable updates
    Sg += (-Sg / tau_gaba +  (C_GABA>0) * (V>Vth)) * dt # GABA
    Sa += (-Sa / tau_ach + (C_ACh>0) * (V>Vth)) * dt # ACh
    Sn += (-Sn / tau_nmda) * dt # NMDA
    
    # reset membrane potential
    S = V>Vth # binary (spike train)
    V[V>Vth] = Vreset # mV (membrane potential)

    return V,Sg,Sa,Sn,S,I_ACh

# run simulation
for i in range(len(t)-1):
    Vs[:,i+1], Sgs[:,:,i+1], Sas[:,:,i+1], Sns[:,:,i+1], Ss[:,i+1], Is[:,i+1] = lif_update(Vs[:,i], Sgs[:,:,i], Sas[:,:,i], Sns[:,:,i], Ss[:,i], 0.5)
```
